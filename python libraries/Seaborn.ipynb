```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # default dataset with seaborn
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') 
#kind is bascically kind of the plot we want. by default it is scatter
```

```python
sns.pairplot(tips,hue='sex') #runs plot for all the numerical data in your data frame
```

```python
sns.rugplot(tips['total_bill']) 
#the difference between is that in rug plot you just draw line for every data point in rug plot
```

```python
#kde is kernal density estimation plot

sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker') 
#hue is to split it by other indicator or column
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
#determines curl density distribution
```

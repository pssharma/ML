```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # default dataset with seaborn
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws = {'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',
          aspect=0.6,size=8)  
# we can make columns on the basis of sex and can add rows for time
```

# GRIDS

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

# Colors and styles

```python
sns.set_style("darkgrid")
sns.despine(left=True,bottom=True)
sns.countplot(x='sex',data=tips)
```

```python
plt.figure(figsize=(6,7.5)) # we can use matplotlib to set the size of seaborn plot
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips) # to use the figure for posters
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
#find matplotlib color maps for palettes
```

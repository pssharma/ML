```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional method

plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()  #only use if you are not in jupyter notebook
```

```python
plt.subplot(1,2,1) #(noof rows, no of cols, the plot number you are reffering to)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object oriented method

fig = plt.figure() #figure object

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width , height 

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')

```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes1.set_title('Larger plot')

axes2.plot(y,x)
axes.set_title('Smaller plot')
```

```python
fig,axes = plt.subplots(nrows =1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))   # size is in inches

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#you can save the maps you produce to any file format jpeg, png ,pdf etc...

fig.savefig('plot.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='squares') #label is for legends
ax.plot(x,x**3,label='cube') #we will get two plots on the same figure


#adding legends

ax.legend(loc=0) #loc is for the location of the legend see documentation, 0 for best location choosen by matplotlib
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color ='red',lw=5,alpha=0.5,ls = '-.',marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='green') #color can take nay arguments like color strings, hex codes rgb

#alpha is for transparency,lw for line width, see documentation for other line styles
```

# Plot range and control over axis

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color ='red',lw=2,alpha=0.5,ls = '-.')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
